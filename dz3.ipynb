```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoic', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

def homework1 (doc):
    num = input('Введите номер документа: \n')
    flag = False
    for key in doc:
        if num in key.values():
            flag = True
            print(key.get('name'))
    if not(flag):
        print('Документ не найден в базе')
    
def homework2 (direct):
    num = input('Введите номер документа: \n')
    flag = False
    for k, v in direct.items():
        if num in v:
            flag = True
            print('Документ хранится на полке:',int(k))
    if not(flag):
        print('Документ не найден в базе')
        
def polka (num,direct):
    for k, v in direct.items():
        if num in v:
            result = k
    return result

def homework3(doc, direct):
    for element in doc:
        print ('№:',element.get('number'), ' тип:', element.get('type'), ' владелец:', element.get('name'), 
               ' полка хранения:', polka(element.get('number'),directories))
        
def homework4(direct):     
    num = input('Введите номер полки: \n')
    if num in direct.keys():
        print('Такая полка уже существует. Текущий перечень полок: ',end="")
        for i in direct.keys():
            print(i,end=" ")
    else:
        direct[num] = ''
        print('Полка добавлена. Текущий перечень полок: ',end="")
        for i in direct.keys():
            print(i,end=" ")

def homework5(direct): 
    num = input('Введите номер полки: \n')
    if num in direct.keys():
        if len(direct[num]) != 0:
            print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:',end="")
            for i in direct.keys():
                print(i,end=" ")
        else:
            del direct[num]
            print('Полка удалена. Текущий перечень полок:',end="")
            for i in direct.keys():
                print(i,end=" ")
    else:
        print('Такой полки не существует. Текущий перечень полок:',end="")
        for i in direct.keys():
            print(i,end=" ")
            
command = '0'
while command != 'q':
    command = input('Введите команду:\n')
    if command == 'p':
        homework1 (documents)
    if command == 's':
        homework2 (directories)
    if command == 'l':
        homework3 (documents,directories)
    if command == 'as':
        homework4 (directories)
    if command == 'ds':
        homework5 (directories)
    
```
